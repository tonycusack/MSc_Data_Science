```python
s = 'OVTK_StimulationId_Label_0'
s1 =  'OVTK_StimulationId_Label_1'

my_list = []
for i in range(1, 10):
    my_list.append(s + str(i))

letter_list = ['A', 'B', 'C', 'D', 'E', 'F']
for i in range(0, 6):
    my_list.append(s + letter_list[i]) 

for i in range(0, 5):
    my_list.append(s1 + str(i)) 
```

```python
word_list = ['date', 'goose', 'spruce', 'knight', 'juice', 
            'moose', 'night', 'queen', 'berry', 'hedgehog', 
            'water', 'daughter', 'gooseberry', 'waterfowl', 
            'wilderness', 'relative', 'watermelon', 'caterpillar', 
            'environment', 'ambassador']
```

```python
lev_distances = {}
cos_similarity = {}
for i in word_list:
    for j in word_list:
        if i != j:
            lev_distances[i + ' ' + j] = lev(i, j)
            cos_similarity[i + ' ' + j] = word2vec.similarity(i, j)
```

```python
cos_similarity = dict(sorted(cos_similarity.items(), key=lambda item: item[1]))
lev_distances = dict(sorted(lev_distances.items(), key=lambda item: item[1]))
```

```python
def expand_session(some_list, session_num):
    if session_num != 9:
        return [c for c, i in zip(some_list, [5] * 25) for _ in range(i)]

    else:
        return [c for c, i in zip(some_list, [256 * 5] * 200) for _ in range(i)]

def label_session(patient_num, session_num, type_of_data, write_to_csv = False):
    try:
        session_to_label = pd.read_csv(f'C:/stims/experiments/{patient_num}/Session {session_num}/' + type_of_data + '.csv') #get what session to label
        if session_num != 9:
            session_labels = word_order[25*(session_num - 1) : 25*session_num] #get the specific labels for that session

            all_epochs = expand_session(session_labels, session_num) #expand to fit the dataframe

            df = session_to_label.copy() #make a copy of the dataframe

            df.loc[:, ['Event Id']] = list(np.array(all_epochs)[np.array(list(df['Epoch']))])

        else: 
            session_labels = word_order
            all_epochs = expand_session(session_labels, session_num)
            df = session_to_label
            df.loc[:, ['Event Id']] = all_epochs


        if write_to_csv:
            os.makedirs(f'C:/stims/experiments/{patient_num}/Labelled Data/Session {session_num}/', exist_ok = True)
            df.to_csv(f'C:/stims/experiments/{patient_num}/Labelled Data/Session {session_num}/' + type_of_data + '.csv', sep=',')
            

        return df

    except: 
        print('Error! Make sure your session_num and type_of_data are valid!')


# for i in range(1, 9):
#     label_session(1, i, 'imagined_speech', True)

# for i in range(1, 9):
#     label_session(1, i, 'inner_speech', True)
```
